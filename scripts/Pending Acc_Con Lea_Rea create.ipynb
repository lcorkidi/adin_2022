```python
def date_value_csv(code, dates):
    date_value_pending = pd.read_csv(join(BASE_DIR, f'_files/development/{datetime.date.today().strftime("%Y-%m-%d")}_date_value_pending.csv'))
    accountable_id_list = []
    for item in dates:
        accountable_id_list.append(code)
    if dates:
        pd.concat([date_value_pending, pd.DataFrame({'accountable_id': accountable_id_list, 'date':dates})])\
            .to_csv(join(BASE_DIR, f'_files/development/{datetime.date.today().strftime("%Y-%m-%d")}_date_value_pending.csv'), float_format='%.0f', na_rep=None, index=False)
```

```python
pending_date_value_dates_df.apply(lambda x: date_value_csv(x.code, x.pending_date_value_dates), axis=1)
```
