```python
import os
from adin.utils.data_credito import IdTypeParse, CompleteNameParse, DateParse, RoleParse, DueAgeParse, DueStateParse, GradeParse, StartValueParse, FeeCountParse
from accountables.utils.accounting_data import ACCOUNT_RECEIPT_PRIORITY

os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

pd.set_option('display.max_rows', 500)
```

```python
objs_df=pd.DataFrame(Lease_Realty_Person.objects.filter(role__in=[1, 2]).values('lease', 'person', 'role', 'person__id_type', 'phone', 'e_mail', 'address'))
objs_df
```

```python
report_df=objs_df.assign(
    Tipo_de_Identificacion=objs_df.person__id_type.apply(lambda x: IdTypeParse(x)),
    Numero_de_Identificacion=objs_df.person.apply(lambda x: x),
    Nombre_Completo=objs_df.person.apply(lambda x: CompleteNameParse(Person.objects.get(pk=x))),
    Situacion_del_Titular=0,
    Fecha_Apertura=objs_df.lease.apply(lambda x: DateParse(Lease_Realty.objects.get(pk=x).doc_date)),
    Fecha_Vencimiento=objs_df.lease.apply(lambda x: DateParse(nextyearlydate(Lease_Realty.objects.get(pk=x).doc_date, datetime.date.today()))),
    Responsable=objs_df.role.apply(lambda x: RoleParse(x)),
    Forma_de_Pago=0,
    Novedad=objs_df.lease.apply(lambda x: DueAgeParse(Charge.pending.accountable_receivable_age_months(Accountable.objects.get(pk=x), ACCOUNT_RECEIPT_PRIORITY))),
    Situacion_Cartera=0,
    Fecha_Situacion_Cartera=objs_df.lease.apply(lambda x: DateParse(Lease_Realty.objects.get(pk=x).doc_date)),
    Estado_de_la_Cuenta=objs_df.lease.apply(lambda x: DueStateParse(Charge.pending.accountable_receivable_age_days(Accountable.objects.get(pk=x), ACCOUNT_RECEIPT_PRIORITY))),
    Fecha_Estado_de_la_Cartera=DateParse(datetime.date.today()),
    Adjetivo='00',
    Fecha_Adjetivo=DateParse(datetime.date.today()),
    Calificacion=objs_df.lease.apply(lambda x: GradeParse(Charge.pending.accountable_receivable_age_months(Accountable.objects.get(pk=x), ACCOUNT_RECEIPT_PRIORITY))),
    Edad_de_Mora=objs_df.lease.apply(lambda x: Charge.pending.accountable_receivable_age_days(Accountable.objects.get(pk=x), ACCOUNT_RECEIPT_PRIORITY)),
    Valor_Inicial=objs_df.lease.apply(lambda x: StartValueParse(Lease_Realty.objects.get(pk=x))),
    Valor_Saldo_Deuda=objs_df.lease.apply(lambda x: Charge.pending.accountable_receivable_sum(Accountable.objects.get(pk=x), ACCOUNT_RECEIPT_PRIORITY)),
    Valor_Disponible=0,
    Valor_Cuota_Mensual=objs_df.lease.apply(lambda x: Accountable.objects.get(pk=x).date_value.exclude(state=0).filter(date__lte=datetime.date.today()).latest('date').value),
    Valor_Saldo_Mora=objs_df.lease.apply(lambda x: Charge.pending.accountable_receivable_sum(Accountable.objects.get(pk=x), ACCOUNT_RECEIPT_PRIORITY)),
    Total_Cuotas=objs_df.lease.apply(lambda x: FeeCountParse(Accountable.objects.get(pk=x)))
    )
report_df
```
