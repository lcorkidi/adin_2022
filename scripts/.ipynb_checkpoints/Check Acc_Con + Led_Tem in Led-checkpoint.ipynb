```python
def Pending_Charge(acc_con, led_tem):
    for cha_tem in led_tem.charges_templates.all():
        if not Charge.objects.filter(
            account=cha_tem.account,
            value=cha_tem.factor.factored_value(acc_con.accountable, acc_con.date, acc_con.value, cha_tem.nature),
            concept=acc_con
        ).exists():
            return False
    return True
```

```python
pending_charge_df=objs_df.assign(pending_charge=objs_df.apply(lambda x:  Pending_Charge(Accountable_Concept.objects.get(pk=x.code), led_tem), axis=1))
pending_charge_df
```

```python
pending_charge_df.apply(lambda x: led_tem.create_ledger(Accountable_Concept.objects.get(pk=x.code), x.date, user) if Accountable_Concept.objects.get(pk=x.code).Pending_Charge(led_tem) else pass, axis=1)
```
