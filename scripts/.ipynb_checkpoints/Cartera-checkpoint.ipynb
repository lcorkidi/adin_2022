```python
per=Person.objects.get(pk=70903810)
per
```

```python
objs_df=pd.DataFrame(Charge.objects.filter(account__in=[13802005], value__gt=0, ledger__third_party=per).order_by('concept__date').values()).drop(['state_change_user_id', 'state_change_date', 'state', 'settled', 'id'], axis=1)
objs_df
```

```python
receipt=-Charge.objects.filter(account__in=[13802005], value__lt=0, ledger__third_party=per).aggregate(receipt=Sum('value'))['receipt']
receipt
```

```python
def DueAge(_dueDate):
    bufferDate=_dueDate
    months=0
    days=0
    while bufferDate < datetime.date.today():
        if bufferDate + relativedelta(months=1) < datetime.date.today():
            months = months +1
        else:
            days=(datetime.date.today() - bufferDate).days
        bufferDate = bufferDate + relativedelta(months=1)
    return f'{months}, {days}'
```

```python
value_sum_df=objs_df.assign(due=objs_df['value'].cumsum())
due_sum_df = value_sum_df.assign(due=value_sum_df.due.apply(lambda x: 0 if x < receipt else x - receipt))
due_df=due_sum_df.assign(due=due_sum_df.apply(lambda x: x.due if x.due < x.value else x.value, axis=1))
filtered_due_df=due_df[due_df['due'] > 0]
age_due_df=filtered_due_df.assign(age=filtered_due_df.concept_id.apply(lambda x: (datetime.date.today() - Accountable_Concept.objects.get(pk=x).date).months))
age_due_df
```
