```python
import os
from os.path import join
from adin.settings import BASE_DIR

from adin.utils.date_progression import nextmonthlydate, nextyearlydate, previousmonthlydate, previousyearlydate

os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

pd.set_option('display.max_rows', 1000)
```

```python
cls=Accountable_Concept
```

```python
code_df = pd.DataFrame(cls.objects.values('code'))
code_df
```

```python
obj_df=code_df.assign(accountable_concept=code_df.code.apply(lambda x: Accountable_Concept.objects.get(pk=x)))
obj_df
```

```python
pen_bil_df=obj_df.assign(
    pending_bill=obj_df.accountable_concept.apply(lambda x: x.Pending_Ledger(x.accountable.accountable_transaction_type.get(transaction_type=x.transaction_type).bill_template)),
    pre_acc_con=obj_df.accountable_concept.apply(lambda x: x.accountable.accountable_concept.exclude(state=0).filter(date__lt=x.date).exists())
)
pen_bil_df
```

```python
from accountables.utils.accounting_data import ACCOUNT_RECEIPT_PRIORITY
```

```python
bil_df=pen_bil_df.assign(pre_due_non=pen_bil_df.apply(lambda x: x.accountable_concept.accountable.accountable_concept.exclude(state=0).filter(date__lt=x.accountable_concept.date).latest('date').ReceivableDueNone(ACCOUNT_RECEIPT_PRIORITY) if x.pre_acc_con else False, axis=1))
bil_df
```

```python
fil_bil_df=bil_df.assign(to_bill=bil_df.apply(lambda x: True if x.pending_bill and (not x.pre_acc_con or (x.pre_acc_con and x.pre_due_non)) else False , axis=1))
fil_bil_df
```
