```python
import os
from os.path import join
from adin.settings import BASE_DIR

from adin.utils.date_progression import nextmonthlydate, nextyearlydate, previousmonthlydate, previousyearlydate

os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

pd.set_option('display.max_rows', 500)
```

```python
per_add_df = pd.read_csv(join(BASE_DIR, f"_files/load/person_address.csv"), keep_default_na=False)
per_add_df
```

```python
def GetAddress(code):
    try:
        return Address.objects.get(code=code)
    except:
        return "None"
```

```python
per_add_df = per_add_df.assign(address=per_add_df.address_id.apply(lambda x: GetAddress(x)))
per_add_df
```
