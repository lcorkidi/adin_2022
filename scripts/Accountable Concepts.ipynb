```python
import os
from os.path import join
from adin.settings import BASE_DIR

from adin.utils.date_progression import nextmonthlydate, nextyearlydate, previousmonthlydate, previousyearlydate

os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

pd.set_option('display.max_rows', 500)
```

```python
cls = Lease_Realty
```

```python
objs_df = pd.DataFrame(cls.objects.values()).drop(['state_change_user_id', 'state_change_date', 'state', 'subclass_id'], axis=1)
errors_report=objs_df.assign(errors=objs_df[f'{cls._meta.pk.name}_id'].apply(lambda x: eval(f'cls.objects.get(pk=x).{cls._meta.pk.name}.get_obj_errors()'))).drop('accountable_ptr_id', axis=1)
errors_report
```

```python
lea_rea_con_data = pd.read_csv(join(BASE_DIR, f"_files/lea_rea_con_data.csv"), keep_default_na=False)
lea_rea_con_data
```

```python
lea_rea_merged=errors_report.merge(lea_rea_con_data[['pk', 'first_concept_date']], left_on='code', right_on='pk').drop('pk', axis=1)
pending_dates=lea_rea_merged\
    .assign(pending_accountable_concept_dates=lea_rea_merged.apply(lambda x: Lease_Realty.objects.get(pk=x.code).pending_accountable_concept_dates(datetime.date(int(x.first_concept_date.split('-')[0]), int(x.first_concept_date.split('-')[1]), int(x.first_concept_date.split('-')[2])) if x.first_concept_date else None), axis=1))\
    .assign(pending_date_value_dates=lea_rea_merged.apply(lambda x: Lease_Realty.objects.get(pk=x.code).pending_date_value_dates(datetime.date(int(x.first_concept_date.split('-')[0]), int(x.first_concept_date.split('-')[1]), int(x.first_concept_date.split('-')[2])) if x.first_concept_date else None), axis=1))
pending_dates
```
