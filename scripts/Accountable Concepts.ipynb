```python
def get_value4date(obj, date):
    if not obj.end_date or obj.end_date >= nextmonthlydate(obj.doc_date, date):
        return int(round(((nextmonthlydate(obj.doc_date, date) - date).days / (nextmonthlydate(obj.doc_date, date) - previousmonthlydate(obj.doc_date, date)).days) * int(obj.date_value.filter(date__lte=item).latest('date').value), 0))
    else:
        return int(round(((obj.end_date - previousmonthlydate(obj.doc_date, date)).days / (nextmonthlydate(obj.doc_date, date) - previousmonthlydate(obj.doc_date, date)).days) * int(obj.date_value.filter(date__lte=item).latest('date').value), 0))
    
```

```python
def pending_date_value_dict(obj, ref_date=None):
    date_value_dict = {}
    for item in obj.pending_accountable_concept_dates(ref_date):
        if obj.date_value.filter(date__lte=item).exists() and obj.date_value.filter(date__lte=item).latest('date').date >= previousyearlydate(obj.doc_date, item):
            date_value_dict[item] = get_value4date(obj, item)        
        else:
            date_value_dict[item] = None        
    return date_value_dict
```

```python
pending_dates.assign(concept_values=pending_dates.apply(lambda x: get_date_value_dict(Lease_Realty.objects.get(pk=x.code), datetime.date(*[int(dt) for dt in x.first_concept_date.split('-')]) if x.first_concept_date else None), axis=1))
```

```python
def date_value_csv(code, dates):
    date_value_pending = pd.read_csv(join(BASE_DIR, f'_files/date_value_pending.csv'))
    accountable_id_list = []
    for item in dates:
        accountable_id_list.append(code)
    if dates:
        pd.concat([date_value_pending, pd.DataFrame({'accountable_id': accountable_id_list, 'date':dates})])\
            .to_csv(join(BASE_DIR, f'_files/date_value_pending.csv'), float_format='%.0f', na_rep=None, index=False)
```
