```python
cls = Date_Value
```

```python
objs_df = pd.DataFrame(cls.objects.values()).drop(['state_change_user_id', 'state_change_date', 'state'], axis=1)
objs_df
```

```python
errors_report=objs_df.assign(errors=objs_df[cls._meta.pk.name].apply(lambda x: cls.objects.get(pk=x).get_obj_errors()))
errors_report
```

```python
children_errors_report=objs_df.assign(errors=objs_df[f'{cls._meta.pk.name}_id'].apply(lambda x: eval(f'cls.objects.get(pk=x).{cls._meta.pk.name}.get_obj_errors()')))
errors_report=children_errors_report
children_errors_report
```

```python
errors_report=errors_report[errors_report['errors'].map(lambda x: len(x) > 0)]
errors_report
```

```python
errors_report.to_csv(join(BASE_DIR, f'_files/date_value_pending.csv'), float_format='%.0f', na_rep=None, index=False)
```
