```python
import os
from adin.utils.data_credito import IdTypeParse, CompleteNameParse, DateParse, RoleParse, DueAgeParse, DueStateParse, GradeParse, StartValue, FeeCount, SettledFeeCount, PendingFeeCount, CityCodeParse, AddressParse
from accountables.utils.accounting_data import ACCOUNT_RECEIPT_PRIORITY

os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

pd.set_option('display.max_rows', 500)
```

```python
objs_df=pd.DataFrame(Lease_Realty_Person.objects.filter(role__in=[1, 2]).values('id', 'lease', 'person', 'role', 'person__id_type'))
objs_df=objs_df.assign(lease=objs_df.lease.apply(lambda x: Lease_Realty.objects.get(pk=x)))
objs_df
```

```python
today=datetime.date.today()
```

```python
report_df=objs_df.assign(
    Tipo_de_Identificacion=objs_df.person__id_type.apply(lambda x: IdTypeParse(x)),
    Numero_de_Identificacion=objs_df.person.apply(lambda x: x),
    Nombre_Completo=objs_df.person.apply(lambda x: CompleteNameParse(Person.objects.get(pk=x))),
    Situacion_del_Titular=0,
    Fecha_Apertura=objs_df.lease.apply(lambda x: DateParse(x.doc_date)),
    Fecha_Vencimiento=objs_df.lease.apply(lambda x: DateParse(nextyearlydate(x.doc_date, today))),
    Responsable=objs_df.role.apply(lambda x: RoleParse(x)),
    Forma_de_Pago=0,
    Novedad=objs_df.lease.apply(lambda x: DueAgeParse(Charge.pending.accountable_receivable_age_months(Accountable.objects.get(pk=x), ACCOUNT_RECEIPT_PRIORITY))),
    Situacion_Cartera=0,
    Fecha_Situacion_Cartera=objs_df.lease.apply(lambda x: DateParse(Lease_Realty.objects.get(pk=x).doc_date)),
    Estado_de_la_Cuenta=objs_df.lease.apply(lambda x: DueStateParse(Charge.pending.accountable_receivable_age_days(x, ACCOUNT_RECEIPT_PRIORITY))),
    Fecha_Estado_de_la_Cartera=DateParse(today),
    Adjetivo='00',
    Fecha_Adjetivo=DateParse(today),
    Calificacion=objs_df.lease.apply(lambda x: GradeParse(Charge.pending.accountable_receivable_age_months(x, ACCOUNT_RECEIPT_PRIORITY))),
    Edad_de_Mora=objs_df.lease.apply(lambda x: Charge.pending.accountable_receivable_age_days(x, ACCOUNT_RECEIPT_PRIORITY)),
    Valor_Inicial=objs_df.lease.apply(lambda x: StartValue(x)),
    Valor_Saldo_Deuda=objs_df.lease.apply(lambda x: Charge.pending.accountable_receivable_sum(x, ACCOUNT_RECEIPT_PRIORITY)),
    Valor_Disponible=0,
    Valor_Cuota_Mensual=objs_df.lease.apply(lambda x: x.date_value.exclude(state=0).filter(date__lte=today).latest('date').value),
    Valor_Saldo_Mora=objs_df.lease.apply(lambda x: Charge.pending.accountable_receivable_sum(x, ACCOUNT_RECEIPT_PRIORITY)),
    Total_Cuotas=objs_df.lease.apply(lambda x: FeeCount(x)),
    Cuotas_Canceladas=objs_df.lease.apply(lambda x: SettledFeeCount(x)),
    Cuotas_en_Mora=objs_df.lease.apply(lambda x: PendingFeeCount(x)),
    Clausula_de_Permanencia=12,
    Fecha_Clausula_de_Permanencia=objs_df.lease.apply(lambda x: DateParse(previousyearlydate(x.doc_date, today))),
    Fecha_Limite_de_Pago=objs_df.lease.apply(lambda x: DateParse(Charge.pending.accountable_receivable_age_start_date(x, ACCOUNT_RECEIPT_PRIORITY))),
    Fecha_de_Pago=objs_df.lease.apply(lambda x: DateParse(Charge.objects.exclude(state=0).filter(concept__accountable=x, ledger__type__abreviation='RC').latest('ledger__date').ledger.date)
                                                         if Charge.objects.exclude(state=0).filter(concept__accountable=x, ledger__type__abreviation='RC').exists()
                                                         else None),
    Ciudad_Correspondencia=objs_df.id.apply(lambda x: Lease_Realty_Person.objects.get(pk=x).address.city),
    Codigo_Dane_Ciudad_Correspondencia=objs_df.id.apply(lambda x: CityCodeParse(Lease_Realty_Person.objects.get(pk=x).address.city)),
    Departamento_de_Correspondencia=objs_df.id.apply(lambda x: Lease_Realty_Person.objects.get(pk=x).address.region),
    Direccion_de_Correspondecia=objs_df.id.apply(lambda x: AddressParse(Lease_Realty_Person.objects.get(pk=x).address)),
    Correo_Electronico=objs_df.id.apply(lambda x: Lease_Realty_Person.objects.get(pk=x).e_mail),
    Celular=objs_df.id.apply(lambda x: Lease_Realty_Person.objects.get(pk=x).phone if Lease_Realty_Person.objects.get(pk=x).phone and Lease_Realty_Person.objects.get(pk=x).phone.type == 1 else None)
    )
report_df
```
